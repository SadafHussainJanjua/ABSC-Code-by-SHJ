```python
explicit_aspects = set()
```

```python
for item in apriori(trans):
        pair = list(item[0])
        for aspect in pair[0].split():
            explicit_aspects.add(aspect)
        if len(pair) > 1:
            for aspect in pair[1].split():
                explicit_aspects.add(aspect)
```
